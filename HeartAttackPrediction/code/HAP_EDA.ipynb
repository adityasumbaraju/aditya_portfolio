```python
from ipynb.fs.full.HAP_DataProcessing import *
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
hap_df = pd.read_csv('C:\BU\portfolio\HeartAttackPrediction\dataset/heart_attack.csv')
data=hap_df
hap_df_eda=hap_df
hap_df.info()
```

```python
d= {1:"Heart Attack",0:"No Heart Attack"}
hap_df_eda.num = data.num.map(d)
s= {1:"Male",0:"Female"}
hap_df_eda.sex = data.sex.map(s)
c ={1:"typical angina",2:"atypical angina",3:"non-anginal pain",4:"asymptomatic"}
hap_df_eda.cp = data.cp.map(c)
```

```python
hap_df_eda.info()
```

```python
ax = sns.countplot(x = 'sex', hue = 'num', palette = 'Set1', data = hap_df_eda)
ax.set(title = 'Patient demographic against Patient Gender', 
       xlabel = 'Gender', ylabel = 'Patients Count')

plt.show()
```

```python
interval = (0,18,35,60,120)
categories = ['Children','Teens','Adult', 'Senior']
hap_df_eda['age'] = pd.cut(hap_df_eda.age, interval, labels = categories)

ax = sns.countplot(x = 'age',  data = hap_df_eda, hue = 'num', palette = 'Set1')

ax.set(xlabel='patient category w.r.t Age', ylabel='Patients Count',
       title="Categorical Distribution w.r.t Age")

plt.show()
```

```python
ax = sns.countplot(x = 'cp', hue = 'num', palette = 'Set2', data = hap_df_eda)
ax.set(title = 'Patient Outcome against Chest Pain Type', 
       xlabel = 'Chest Pain Type', ylabel = 'Patients Count')

plt.show()
```

```python
fig = plt.figure(figsize=(18,18))
gs = fig.add_gridspec(5,2)
ax = fig.add_subplot(gs[2,1])

# Chol
ax.grid(color='#000000', linestyle=':', axis='y', zorder=0,  dashes=(1,5))
sns.kdeplot(ax=ax, data=hap_df_eda, x='chol',hue="num", fill=True,palette=["#8000ff","#da8829"], alpha=.5, linewidth=0)
ax.set_xlabel("")
ax.set_ylabel("")
```

```python
s= {1:"Diabetes",0:"Normal"}
hap_df_eda.fbs = hap_df_eda.fbs.map(s)
ax = sns.countplot(x = 'fbs', hue = 'num', palette = 'Set1', data = hap_df_eda)
ax.set(title = 'Patient demographics against Fasting Blood Sugar', 
       xlabel = 'Fasting Blood Sugar', ylabel = 'Patients Count')

plt.show()
```

```python
plt.figure(figsize=(20,15))
sns.displot(hap_df_eda["trestbps"])
plt.title("Distribution of blood pressure",fontsize=20)
plt.xlabel("BLOOD PRESSURE",fontsize=20)
plt.ylabel("COUNT",fontsize=20)
plt.show()
```

```python
s= {1:"Yes",0:"No"}
hap_df_eda.exang = hap_df_eda.exang.map(s)
ax = sns.countplot(x = 'exang', hue = 'num', palette = 'Set3', data = hap_df_eda)
ax.set(title = 'Patient Outcome against Exercise Induced Angina', 
       xlabel = 'Exercise Induced Angina', ylabel = 'Patients Count')
```

```python
ax = sns.countplot(x = 'thal', hue = 'num', palette = 'Set1', data = hap_df_eda)
ax.set(title = 'Patient Outcome against thal', 
       xlabel = 'thal', ylabel = 'Patients Count')

plt.show()
```

```python
pd.crosstab(hap_df_eda.age,hap_df_eda.num).plot(kind="bar",figsize=(20,6))
plt.title('Heart Disease Frequency w.r.t Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
